### Feature Engineering

```python
#converting categorical features to numerical values
categor_col
```

```python
# map "yes" to 0 and "no" to 1
df['default']=df['default'].map({'yes':1,'no':0})
```

```python
df['housing']=df['housing'].map({'yes':1,'no':0})
```

```python
df['subscribed']=df['subscribed'].map({'yes':1,'no':0})
```

```python
df['loan']=df['loan'].map({'yes':1,'no':0})
```

```python
#one hot encoding of the job column
dfJob=pd.get_dummies(df['job'])
dfJob
```

```python
# Concatenate the original dataframe with dataframe created. 
df=pd.concat([df,dfJob],axis=1)
df
```

```python
#one hot encoding of the marital column
dfMarital=pd.get_dummies(df['marital'])
dfMarital
```

```python
# Concatenate the original dataframe with dataframe created 
df=pd.concat([df,dfMarital],axis=1)
df
```

```python
#one hot encoding of the education column
dfEducation=pd.get_dummies(df['education'],prefix="education")
dfEducation
```

```python
# Concatenate the original dataframe with dataframe created 
df=pd.concat([df,dfEducation],axis=1)
df
```

```python
#one hot encoding of the month column
dfMonth=pd.get_dummies(df['month'])
dfMonth
```

```python
# Concatenate the original dataframe with dataframe created 
df=pd.concat([df,dfMonth],axis=1)
df
```

```python
#one hot encoding of the contact column
dfContact=pd.get_dummies(df['contact'],prefix="contact")
dfContact
```

```python
# Concatenate the original dataframe with dataframe created 
df=pd.concat([df,dfContact],axis=1)
df
```

```python
#one hot encoding of the outcome column
dfOutcome=pd.get_dummies(df['poutcome'],prefix="outcome")
dfOutcome
```

```python
# Concatenate the original dataframe with dataframe created 
df=pd.concat([df,dfOutcome],axis=1)
df
```

```python
df.columns
```

```python
# Create X with all independent variable
X=df[['age','default', 'balance', 'housing',
       'loan','day','duration', 'campaign', 'pdays',
       'previous','admin.', 'blue-collar',
       'entrepreneur', 'housemaid', 'management', 'retired', 'self-employed',
       'services', 'student', 'technician', 'unemployed', 'unknown',
       'divorced', 'married', 'single', 'education_primary',
       'education_secondary', 'education_tertiary', 'education_unknown', 'apr',
       'aug', 'dec', 'feb', 'jan', 'jul', 'jun', 'mar', 'may', 'nov', 'oct',
       'sep', 'contact_cellular', 'contact_telephone', 'contact_unknown',
       'outcome_failure', 'outcome_other', 'outcome_success',
       'outcome_unknown']]
X
```

```python
# Create y with only target variable(charges)
y=df['subscribed']
y
```

```python
#Feature scaling
ss=StandardScaler()
```

```python
# Apply standard scaler on only relevant columns
X[['age','balance','day','duration','campaign','pdays', 'previous']]=ss.fit_transform(X[['age','balance','day','duration','campaign','pdays', 'previous']])
X
```

#### Oversampling the data as subscribed values are not balanced

```python
y.value_counts()
```

```python
sm=SMOTE()
X,y=sm.fit_resample(X,y)
```

```python
y.value_counts()
```

### Model Building

```python
#Split the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=80)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

#### Logistic Regression

```python
# Fit the Logistic Regression model
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
# Predict on test data
y_test_pred=lr.predict(X_test)
y_test_pred
```

```python
# Print the confusion matrix which has the TruePositive, FalsePositive, FalseNegative and TrueNegative counts 
confusion_matrix(y_test,y_test_pred)
```

```python
# Print the accuracy on test data
accuracy_score(y_test,y_test_pred)
```

#### Random Forest

```python
# Fit the Random Forest Classifier model
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
# Predict on test data
y_test_pred=rf.predict(X_test)
y_test_pred
```

```python
# Print the confusion matrix on test data
confusion_matrix(y_test,y_test_pred)
```

```python
# Print the accuracy on test data
accuracy_score(y_test,y_test_pred)
```

The accuracy of Random Forest model is 95%.
The Random Forest model is better than Logistic Regression model.

```python
# Create X with all independent variable
XTest=dfTest[['age','default', 'balance', 'housing',
       'loan','day','duration', 'campaign', 'pdays',
       'previous','admin.', 'blue-collar',
       'entrepreneur', 'housemaid', 'management', 'retired', 'self-employed',
       'services', 'student', 'technician', 'unemployed', 'unknown',
       'divorced', 'married', 'single', 'education_primary',
       'education_secondary', 'education_tertiary', 'education_unknown', 'apr',
       'aug', 'dec', 'feb', 'jan', 'jul', 'jun', 'mar', 'may', 'nov', 'oct',
       'sep', 'contact_cellular', 'contact_telephone', 'contact_unknown',
       'outcome_failure', 'outcome_other', 'outcome_success',
       'outcome_unknown']]
XTest
```

```python
# Apply standard scaler on only relevant columns
XTest[['age','balance','day','duration','campaign','pdays', 'previous']]=ss.transform(XTest[['age','balance','day','duration','campaign','pdays', 'previous']])
XTest
```

```python
# Predict on test data
rf.predict(XTest)
```
