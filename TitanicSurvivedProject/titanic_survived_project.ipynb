### Feature Engineering

```python
#converting categorical features to numerical values
categorical_cols
```

```python
# map "male" to 0 and "female" to 1
df['Sex']=df['Sex'].map({'male':0,'female':1})
```

```python
#one hot encoding of the region column
dfEmbarked=pd.get_dummies(df['Embarked'],prefix="Embarked")
dfEmbarked
```

```python
# Concatenate the original dataframe with dataframe created for embarked values, to concatenate two dataframe column-wise we have used axis=1
df=pd.concat([df,dfEmbarked],axis=1)
df
```

```python
df.columns
```

```python
# Create X with all independent variable
X=df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare','Embarked_C', 'Embarked_Q', 'Embarked_S']]
X
```

```python
# Create y with only target variable(charges)
y=df["Survived"]
y
```

```python
#Feature scaling
ss=StandardScaler()
```

```python
# Apply standard scaler on only relevant columns
X[['Age','Fare']]=ss.fit_transform(X[['Age','Fare',]])
X
```

The classes are not that imbalanced so class balancing is not required.

### Model Building

```python
#Split the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=80)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

#### Logistic Regression

```python
# Fit the Logistic Regression model
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
# Predict on test data
y_test_pred=lr.predict(X_test)
y_test_pred
```

```python
# Print the confusion matrix which has the TruePositive, FalsePositive, FalseNegative and TrueNegative counts 
confusion_matrix(y_test,y_test_pred)
```

```python
# Print the accuracy on test data
accuracy_score(y_test,y_test_pred)
```

The accuracy of Logistic Regression model is 78%

#### Random Forest

```python
# Fit the Random Forest Classifier model
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
# Predict on test data
y_test_pred=rf.predict(X_test)
y_test_pred
```

```python
# Print the confusion matrix on test data
confusion_matrix(y_test,y_test_pred)
```

```python
# Print the accuracy on test data
accuracy_score(y_test,y_test_pred)
```

The accuracy of Random Forest model is 82%.
The Random Forest model is better than Logistic Regression model.
