#### Handling Null values

```python
missingValuesColumns=('CY-105', 'HS-105/12', 'MT-111','CS-106', 'EL-102', 'EE-119', 'ME-107', 'CS-107', 'HS-205/20', 'MT-222', 'EE-222', 'MT-224', 'CS-210', 'CS-211', 'CS-203', 'CS-214', 'EE-217', 'CS-212', 'CS-215', 'MT-331', 'EF-303', 'HS-304', 'CS-301','CS-302', 'TC-383', 'MT-442', 'EL-332', 'CS-318', 'CS-306', 'CS-312','CS-317', 'CS-403', 'CS-421', 'CS-406', 'CS-414', 'CS-419', 'CS-423','CS-412')
```

```python
missingValuesColumns
```

```python
#check value count for missingValuesColumns
for i in missingValuesColumns:
    print(df[i].value_counts())
    print("\n")
```

```python
df['CY-105'].value_counts().index[0]
```

```python
# filling all missing values with their respective modes
for i in missingValuesColumns:
    df[i]=df[i].fillna(df[i].value_counts().index[0])
```

```python
df.info()
```

```python
#dropping the following column
df.drop(columns=['Seat No.'],inplace=True)
```

### Univariate Analysis

```python
df.columns
```

```python
# Prepare list of columns names which contain numerical and categorical values
numerical_cols=['CGPA']
categorical_cols=df.columns[0:-1].tolist()
```

```python
# As the grades are ordinal mapping it to a numeric value
for i in categorical_cols:
    df[i]=df[i].map({'F':0,'W':0,'WU':0,'I':0,'D-':1,'D':2,'D+':3,'C-':4,'C':5,'C+':6,'B-':7,'B':8,'B+':9,'A-':10,'A':11,'A+':12})
    print(df[i].value_counts())
    print("\n")
```

```python
df.info()
```

```python
#Display distribution of all numrical_cols 
for i in numerical_cols+categorical_cols:
    sns.distplot(df[i])
    plt.show()
```

### Bivariate Analysis

```python
# For all categorical column compare the values ranges of charges(target variable)
for i in categorical_cols:
    sns.scatterplot(y=df['CGPA'],x=df[i])
    plt.show()
```

### Multivariate Analysis

```python
# Plot a heatmap using dataframe correlation values
plt.figure(figsize=(15,9))
sns.heatmap(df.corr())
```

Looks like the numerical columns has very low correlation with each other.

## Feature Engineering

```python
# Create X with all independent variable
X=df[categorical_cols]
X
```

```python
# Create y with only target variable(charges)
y=df['CGPA']
y
```

### Model Building

```python
#Split the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=80)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

#### Linear Regression

```python
# Fit the Linear Regression model
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
# Predict on test data
y_test_pred=lr.predict(X_test)
y_test_pred
```

```python
# Display the mean squared error on test data
mean_squared_error(y_test,y_test_pred)
```

```python
# Display the R2 score on test data
r2_score(y_test,y_test_pred)
```

```python
# Plot actual value vs predicted value to visually see how good the predictions are
sns.scatterplot(x=y_test,y=y_test_pred)
```

#### Random Forest

```python
# Fit the Random Forest Regressor model
rf=RandomForestRegressor()
rf.fit(X_train,y_train)
```

```python
# Predict on test data
y_test_pred=rf.predict(X_test)
y_test_pred
```

```python
# Display the mean squared error on test data
mean_squared_error(y_test,y_test_pred)
```

```python
# Display the R2 score on test data
r2_score(y_test,y_test_pred)
```

```python
sns.scatterplot(x=y_test,y=y_test_pred)
```

R2 score using linear regression model is 0.94 and for Random forest is 0.96. The visual representation show the predicted value are closer to the actual values in lot of datapoints. The Random Forest model is better than Linear Regression model.
